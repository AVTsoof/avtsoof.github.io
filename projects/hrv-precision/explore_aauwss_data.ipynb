```python
import pickle
from pathlib import Path
from typing import Any

import numpy as np
import plotly.graph_objects as go
import pandas as pd
from plotly.subplots import make_subplots


from avtsoof.common_utils import data_dir, DOCS
```

```python
ecg_example_path = data_dir("AAUWSS") / "aligned_sleep_data_set" / "ecg" / "subject_01_ecg.pkl"
assert ecg_example_path.exists(), f"Missing {ecg_example_path}"
```

```python
ppg_example_path = data_dir("AAUWSS") / "aligned_sleep_data_set" / "ppg" / "subject_01_ppg.pkl"
assert ppg_example_path.exists(), f"Missing {ppg_example_path}"
```

```python
with open(ecg_example_path, "rb") as f:
    ecg_example_data = pickle.load(f)
```

```python
with open(ppg_example_path, "rb") as f:
    ppg_example_data = pickle.load(f)
```

```python
data = {
    "ecg": ecg_example_data,
    "ppg": ppg_example_data,
}
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(y=data["ecg"][0]))
fig.add_trace(go.Scatter(y=data["ppg"][0]))
fig.show()
```

```python
# normalize all data between -1 and 1
num_epochs = len(data["ecg"])
normalized_data: dict[str, dict[int, pd.DataFrame]] = {"ecg": {}, "ppg": {}}
for epoch in range(num_epochs):
    normalized_data["ecg"][epoch] = 2 * (data["ecg"][epoch] - np.min(data["ecg"][epoch])) / (np.max(data["ecg"][epoch]) - np.min(data["ecg"][epoch])) - 1
    normalized_data["ppg"][epoch] = 2 * (data["ppg"][epoch] - np.min(data["ppg"][epoch])) / (np.max(data["ppg"][epoch]) - np.min(data["ppg"][epoch])) - 1
```

```python
# plot normalized data
fig = go.Figure()
fig.add_trace(go.Scatter(y=normalized_data["ecg"][0]))
fig.add_trace(go.Scatter(y=normalized_data["ppg"][0]))
fig.show()
```

```python
# merge epochs into a new df
epochs_to_merge = 10
merged_ecg = np.concatenate([normalized_data["ecg"][i] for i in range(epochs_to_merge)])
merged_ppg = np.concatenate([normalized_data["ppg"][i] for i in range(epochs_to_merge)])

# plot merged ecg data
fig = go.Figure()
fig.add_trace(go.Scatter(y=merged_ecg))
fig.add_trace(go.Scatter(y=merged_ppg))
fig.show()
```
